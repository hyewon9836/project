```python
import pandas as pd
```

```python
data=pd.read_csv('./data/Lyrics_remove.csv',encoding='utf-8')
data
```

```python
f = open("stopword.txt", "rt",encoding='utf-8')
stopword = []
while True:
    line = f.readline().strip()
    if not line: break
    stopword.append(line)
```

```python
lst=[]
for lyric in data['Lyrics']:
    lyric=lyric.split(' ')
    sub=[]

    sub=[word for word in lyric if not word in stopword]
    lst.append(sub)
print(len(lst))
```

```python
lyrics=[]
for word_lst in lst:
    sen=''
    sen=' '.join(word_lst)
    lyrics.append(sen)
print(len(lyrics))
```

```python
data['Lyrics']=lyrics
data
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

word_lst=sum(lst, [])
wordcloud=WordCloud(font_path='./NanumGothic.ttf').generate(' '.join(word_lst))

plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
data.to_csv('./data/Lyrics_final.csv',encoding='utf-8-sig',index=False)
```
