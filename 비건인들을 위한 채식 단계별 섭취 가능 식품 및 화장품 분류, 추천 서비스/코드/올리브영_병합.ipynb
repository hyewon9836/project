```python
import pandas as pd
import re
```

```python
data1=pd.read_csv('./data/olive0_pre1.csv',encoding='utf-8')
data2=pd.read_csv('./data/olive0_pre2.csv',encoding='utf-8')
print(data1.shape,data2.shape)
```

```python
data1.head()
```

```python
text_lst=[]
for text in data1['text']:
    text=re.sub('[^a-zA-Z가-힣()%0-9,]','',text)
    text_lst.append(text)
data1['text']=text_lst
```

```python
data2.head()
```

```python
lst=data2['text'].tolist()
```

```python
result=[]
for text in lst:
    text_lst=[]
    for word in text.split(','):
        if word=='':
            pass

        if '정선분' in word and len(word)>4:
            word='정성분'

        if '로션' in word and len(word)>4:
            word=''
        
        if '크림' in word and len(word)>3:
            word=''
        
        if '성분' in word and len(word)>4:
            word=''
        
        if 'ML' in word and len(word)>3:
            word=''

        if '표기' in word:
            word=''

        if '&' in word:
            word=''

        if '1제' in word:
            word=''

        if '클렌징' in word:
            word=''
        
        if len(word)>0:
            text_lst.append(word)

    result.append(','.join(text_lst))
result[:2]
```

```python
text_lst=[]
for text in result:
    text=re.sub('[^a-zA-Z가-힣()%0-9,]','',text)
    text_lst.append(text)
```

```python
data2['text']=text_lst
data2.head()
```

```python
# 중복 성분 제거
result=[]
for lst in data2['text']:
    text=lst.split(',')
    text=list(set(text))
    result.append(','.join(text))
```

```python
data2['text']=result
data2
```

```python
# 중복행 제거
data2=data2.drop_duplicates(['brand','product'],keep='first')
data2
```

```python
olive0=pd.concat([data2,data1])
olive0=olive0.reset_index(drop=True)
olive0
```

```python
vegan=pd.read_csv('./data/olive0_vegan.csv',encoding='utf-8')
vegan
```

```python
df=pd.merge(olive0,vegan,how='left')
df['vegan'].fillna('0',inplace=True)
df
```

```python
df.to_csv('./data/olive0.csv',encoding='utf-8-sig',index=False)
```
