```python
import pandas as pd
import re
```

## 데이터 확인

```python
df=pd.read_csv('./data/olive0.csv',encoding='utf-8')
df
```

```python
# 괄호제거
def hasNumber(stringVal):
    return any(elem.isdigit() for elem in stringVal)

text_lst=[]
for text in df['text'].tolist():
    sen=''
    for word in text.split(','):
        if '(' in word:
            word=word.split('(')[1]
        if ')' in word:
            word=word.split(')')[0]
        if len(word)>0 and hasNumber(word)==0:
            sen=sen+','+word
    sen=sen[1:]
    text_lst.append(sen)
print(len(text_lst))

df['text']=text_lst
df
```

## 분류

```python
# 비건카테고리 성분리스트
vegan_lst=[]
for sen in df.loc[df['vegan']==1.0,'text'].tolist():
    for word in sen.split(','):
        vegan_lst.append(word)
vegan_lst=list(set(vegan_lst))
```

```python
# 동물카테고리 불러오기
with open("./list/animal_1.txt", "r") as file:
    lst = file.readlines()

animal_lst=[]
for word in lst:
    word=word.replace('\n','')
    word=word.replace(' ','')
    if len(word)>0:
        animal_lst.append(word)
```

```python
# 분류 체크리스트
check_lst=list(set(animal_lst)-set(vegan_lst))
check_lst
```

```python
# 분류 체크리스트 추가항목
check_lst.append('우유')
check_lst.append('글라이코젠')
```

```python
with open('./list/check_lst.txt','w',encoding='UTF-8') as f:
    for word in check_lst:
        f.write(word+'\n')
```

```python
df['class']='no animal'
df.head(1)
```

```python
# 동/식물 분류
def classify(lst):
    for word in lst.split(','):
        word=word.replace(' ','')
        if word in check_lst:
            ingredient_lst.append(word)
            return 'animal'

ingredient_lst=[]
df['class']=df['text'].apply(classify)
df['class'].fillna('not animal',inplace=True)
df.head()
```

```python
# 동물성 화장품
df_animal=df[df['class']=='animal']
df_animal=df_animal[['brand','product','text','cate']].copy()
df_animal['animal']=1
df_animal
```

```python
# 식물성 화장품
df_plant=df[df['class']=='not animal']
df_plant=df_plant[['brand','product','text','cate']].copy()
df_plant['animal']=0
df_plant
```

```python
# 올리브영 최종본
olive0=pd.concat([df_animal,df_plant])
olive0.reset_index(drop=True,inplace=True)
olive0
```

```python
olive0.to_csv('./data/olive0_classify.csv',encoding='utf-8-sig',index=False)
```
