### 수입식품정보마루

```python
data1=pd.read_excel('./data/식품원료_목록_식물.xlsx')
data1=data1[['원재료','이명 또는 영명','생약명']]
data1
```

```python
data1.fillna('',inplace=True)
```

```python
list1_1=data1['원재료'].tolist()
list1_1
```

```python
lst=data1['이명 또는 영명'].tolist()
lst[:10]
```

```python
list1_2=[]
for ing in lst:
    if ing=='-':
        continue
    elif ',' in ing:
        for word in ing.split(','):
            list1_2.append(word.strip())
    else:
        list1_2.append(ing)
list1_2[:10]
```

```python
lst=data1['생약명'].tolist()
lst
```

```python
list1_3=[]
for ing in lst:
    if ing=='':
        continue
    ing=re.sub('[()]',',',ing)
    if ',' in ing:
        for word in ing.split(','):
            if word =='':
                continue
            list1_3.append(word.strip())
    else:
        list1_3.append(ing.strip())
list1_3
```

```python
list1=list1_1+list1_2+list1_3
print(len(list1))
print(list1[:50])
```

### 식품안전나라_수입식품정보

```python
df=pd.read_excel('./data/식품안전_수입식품.xlsx')
df=df[1:]
df=df[['원재료','이명 또는 영명','생약명','분류']]
df
```

### 육류

```python
meat=[]

# docs food meat
with open("./list/식품_고기.txt", "r") as file:
    lst1 = file.readlines()

for lst in lst1:
    for word in lst.split(','):
        word=word.replace('\n','')
        meat.append(word.strip())

# docs cosmetic meat
with open("./list/화장품_고기.txt", "r") as file:
    lst2 = file.readlines()

for lst in lst2:
    for word in lst.split(','):
        word=word.replace('\n','')
        meat.append(word.strip())
```

```python
df_meat=df[df['분류']=='육류']
df_meat
```

```python
list1=df_meat['원재료'].tolist()
for ing in list1:
    if '(' in ing:
        ing=ing.replace('(',',')
        ing=ing.replace(')','')
    if ',' in ing:
        for word in ing.split(','):
            meat.append(word.strip())
    else:
        meat.append(ing)
```

```python
list2=df_meat['이명 또는 영명'].tolist()
for ing in list2:
    if ing=='-':
        continue
    else:
        if ',' in ing:
            for word in ing.split(','):
                meat.append(word.strip())
```

```python
list3=df_meat['생약명'].tolist()
list3
```

```python
meat=list(set(meat))
meat=meat[1:]
meat
```

```python
with open('./list/meat.txt','w',encoding='UTF-8') as f:
    for word in meat:
        f.write(word+'\n')
```

### 해산물

```python
sea=[]

# docs food seafood
with open("./list/식품_어류.txt", "r") as file:
    lst1 = file.readlines()

for lst in lst1:
    for word in lst.split(','):
        word=word.replace('\n','')
        sea.append(word.strip())

# docs cosmetic seafood
with open("./list/화장품_어류.txt", "r") as file:
    lst2 = file.readlines()

for lst in lst2:
    for word in lst.split(','):
        sea.append(word.strip())
```

```python
df_sea=df[df['분류']=='해산물']
df_sea
```

```python
list1=df_sea['원재료'].tolist()

for ing in list1:
    if '(' in ing:
        ing=ing.replace('(',',')
        ing=ing.replace(')','')
    if ',' in ing:
        for word in ing.split(','):
            sea.append(word.strip())
    else:
        sea.append(ing)
```

```python
list2=df_sea['이명 또는 영명'].tolist()

for ing in list2:
    if ing=='-':
        continue
    else:
        if ',' in ing:
            for word in ing.split(','):
                sea.append(word.strip())
```

```python
list3=df_sea['생약명'].tolist()
for word in list3:
    if word=='-':
        continue
    else:
        sea.append(word)
```

```python
sea=list(set(sea))
sea=sea[0:1]+sea[2:]
sea
```

```python
with open('./list/sea.txt','w',encoding='UTF-8') as f:
    for word in sea:
        f.write(word+'\n')
```
