```python
import pandas as pd
import re
```

```python
df=pd.read_csv('./data/kully_f.csv',encoding='utf-8')
df
```

```python
df['allergy'].tolist()
```

```python
def hasNumber(stringVal):
    return any(elem.isdigit() for elem in stringVal)
```

```python
result_lst=[]
for text in df['allergy']:
        
    # [] 내용 제거
    cnt=0
    try:
        while(1):
            start_idx=re.search('[',text).start()
            end_idx=re.search(']',text).end()
            text=text[:start_idx+1]+text[end_idx+1:]
            cnt+=1
            if cnt>4:
                break
    except:
        pass

    result2=''
    for word in text.split('\n'):
        
        try:
            idx=re.search(']',word).start()
            word=word[idx+1:]
        except:
            pass

        # 기타 처리
        word=word.replace(' ','')
        word=word.replace(':',',')
        word=word.replace('함유식품','')
        word=word.replace('함유','')
        word=word.replace('이제품은','')
        word=word.replace('본제품은','')


        # 함께제조 삭제
        try:
            start_idx=re.search('을사용',word).start()
            word=word[:start_idx]
        except:
            pass

        try:
            start_idx=re.search('와같은',word).start()
            word=word[:start_idx]
        except:
            pass

        try:
            start_idx=re.search('를사용',word).start()
            word=word[:start_idx]
        except:
            pass

        try:
            start_idx=re.search('과같은',word).start()
            word=word[:start_idx]
        except:
            pass
        
        if '상품별' in word:
            word=''

        ## ,로 다시 나눠서 숫자 포함되어 있으면 삭제
        result=''
        for ing in word.split(','):
            if hasNumber(ing)==True:
                pass
            else:
                result+=ing+','

        result2+=result+','

    result_lst.append(result2)
```

```python
len(result_lst)
```

```python
lst=[]
for text in result_lst:
    text=re.sub('[^a-zA-Z가-힣,()]','',text)
    text=re.sub(',+',',',text)

    if len(text)==0:
        text='없음'
    if text[0]==',':
        text=text[1:]
    
    if len(text)==0:
        text='없음'
    if text[-1]==',':
        text=text[:len(text)-1]

    if '별도' in text:
        text='없음'

    text=text.replace('(공통)','')
    
    lst.append(text)
lst
```

```python
import re
result=[]
for idx,text in enumerate(lst):
    text=text.replace('혼입가능','')
    text=text.replace('(',',')
    text=text.replace(')',',')
    text=re.sub(',+',',',text)
    result.append(text)
result
```

```python
df['allergy']=result
df
```

```python
df.isnull().sum()
```

```python
df.to_csv('./data/kurly_final.csv',encoding='utf-8-sig',index=False)
```

```python
import pandas as pd
```

```python
df=pd.read_csv('./data/kurly_final.csv',encoding='utf-8')
vegan=pd.read_csv('./data/크롤링/kurly_vegan.csv',encoding='utf-8')
```

```python
print(df.shape)
df.head()
```

```python
df=df.drop_duplicates(['brand','item'],keep='first')
df
```

```python
vegan.rename(columns = {'product' : 'item'}, inplace = True)
vegan['vegan']='1'
vegan
```

```python
data=pd.merge(df,vegan,how='left')
data['vegan'].fillna('0',inplace=True)
data
```

```python
data.loc[data['item'].str.contains('비건'),'vegan']='1'
data.loc[data['item'].str.contains('식물성'),'vegan']='1'
data.loc[data['item'].str.contains('템페'),'vegan']='1'
data
```

```python
data.to_csv('./data/kurly.csv',encoding='utf-8-sig',index=False)
```
