## 남성 카테고리 크롤링

### 남성 > 스킨케어

```python
url='https://www.oliveyoung.co.kr/store/display/getMCategoryList.do?dispCatNo=100000100070007&isLoginCnt=2&aShowCnt=0&bShowCnt=0&cShowCnt=0&trackingCd=Cat100000100070007_MID'

driver=webdriver.Chrome()
driver.get(url)
act=ActionChains(driver)
```

```python
text_lst=[]
brand_name=[]
product_name=[]
page=0
for page in range(0,10): # 페이지 for문
    for line in range(8,14): # 행 for문
        for i in range(4):
            # 상품 클릭
            driver.find_elements(By.CSS_SELECTOR,'#Contents > ul:nth-child('+str(line)+') > li')[i].click()
            time.sleep(0.5)

            # 브랜드 이름, 상품 이름 크롤링
            html=driver.page_source
            soup=BeautifulSoup(html,'html.parser')
            brand_name.append(soup.select('div.prd_info > p')[0].text)
            product_name.append(soup.select('div.prd_info > p')[1].text)

            # 구매정보 클릭 및 성분 크롤링
            driver.find_elements(By.CSS_SELECTOR,'#buyInfo')[0].click()
            time.sleep(0.5)
            html=driver.page_source
            soup=BeautifulSoup(html,'html.parser')
            text_lst.append(soup.select('div > dl > dd')[8])

            # 이전 화면으로 이동
            driver.back()

            # 다음 페이지로 이동
            time.sleep(0.5)
            driver.find_elements(By.CSS_SELECTOR,'#Container > div.pageing > a')[page].click()
    print('page',page)
```

```python
len(text_lst),len(product_name),len(brand_name)
```

```python
import pandas as pd
data=pd.DataFrame({'brand':brand_name,'product':product_name,'text':text_lst})
data
```

```python
data.to_csv('./data/olive0_1.csv',encoding='utf-8-sig',index=False)
```
