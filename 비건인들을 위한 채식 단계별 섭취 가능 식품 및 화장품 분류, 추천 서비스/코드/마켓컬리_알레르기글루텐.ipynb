```python
import pandas as pd
import re
```

```python
kurly=pd.read_csv('./data/kurly_classify.csv')
kurly['gluten']=''
kurly
```

```python
glu=['글루테닌','글리아딘','보리','귀리','밀','박력분','강력분','중력분','밀가루']

def gluten_ckeck(ingredient):
    gluten_lst=[]
    for word in ingredient.split(','):
        if word in glu:
            gluten_lst.append(word)
    if len(gluten_lst)>0:
        return ','.join(gluten_lst)
    else:
        return '발견사항 없음'

kurly['gluten']=kurly['ingredient'].apply(gluten_ckeck)
kurly
```

```python
kurly.to_csv('./data/kurly_classify_allergy_gluten.csv',encoding='utf-8-sig',index=False)
```
