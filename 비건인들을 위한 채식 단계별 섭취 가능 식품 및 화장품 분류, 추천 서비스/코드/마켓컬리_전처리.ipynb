```python
import pandas as pd
```

```python
df1 = pd.read_csv("C:\propro\kurly\kurly1.csv")
df2 = pd.read_csv("C:\propro\kurly\kurly2.csv")
```

```python
df2['category'] = '음료/우유/커피/차'
df2 = df2[ ['category','item','allergy','img'] ]
df2.head()
```

```python
df1['category'] = '면/앙념/오일'
df1 = df1[ ['category','item','allergy','img'] ]
df1.head()
```

```python
#파일 아래로 추가
merged_df = pd.concat([df1, df2])
merged_df.tail(10)
```

```python
temp_df = pd.DataFrame(merged_df['item'].str.split(']').tolist()).rename(columns={0:"brand", 1:"item"})
temp_df.head()
```

```python
category = merged_df['category']
brand = temp_df['brand'].str.replace('[','')
item = temp_df['item']
allergy = merged_df['allergy']
img = merged_df['img']
```

```python
category = category.reset_index()
brand = brand.reset_index()
item = item.reset_index()
allergy = allergy.reset_index()
img = img.reset_index()
df2 = pd.concat([category,brand,item,allergy,img], axis=1)
df2.columns
```

```python
df2 =df2.drop(['index'],axis=1)
df2.tail()
```

```python
#파일 저장
pd.set_option('display.max_rows', None)
df2
df2.to_csv('C:\propro\kurly\souce_drink.csv',encoding='utf-8-sig',index=False)
```
